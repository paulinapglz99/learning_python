```python
from random import randint

# Generar valor aleatorio de la compu
manoCompu = randint(0, 4)

# Preguntar al usuario su tirada o su mano
print("Escribe una tirada para jugar piedra, papel, tijeras, lagarto o Spock")
print("0 -> piedra")
print("1 -> papel")
print("2 -> tijeras")
print("3 -> lagarto")
print("4 -> Spock")

manoUsuario = int(input())

if manoUsuario == 0 or manoUsuario == 1 or manoUsuario == 2 or manoUsuario == 3 or manoUsuario == 4:
    if manoUsuario == manoCompu:
        print("EMPATE")
    else:
        if manoUsuario == 0:  # Usuario elige piedra
            if manoCompu == 1:  # Papel gana a piedra
                print("Gana compu")
            elif manoCompu == 2:  # Piedra gana a tijeras
                print("Gana usuario")
            elif manoCompu == 3:  # Piedra gana a lagarto
                print("Gana usuario")
            else:  # Spock gana a piedra
                print("Gana compu")
        
        elif manoUsuario == 1:  # Usuario elige papel
            if manoCompu == 0:  # Papel gana a piedra
                print("Gana usuario")
            elif manoCompu == 2:  # Tijeras gana a papel
                print("Gana compu")
            elif manoCompu == 3:  # Lagarto gana a papel
                print("Gana compu")
            else:  # Papel gana a Spock
                print("Gana usuario")
        
        elif manoUsuario == 2:  # Usuario elige tijeras
            if manoCompu == 0:  # Piedra gana a tijeras
                print("Gana compu")
            elif manoCompu == 1:  # Tijeras gana a papel
                print("Gana usuario")
            elif manoCompu == 3:  # Tijeras gana a lagarto
                print("Gana usuario")
            else:  # Spock gana a tijeras
                print("Gana compu")
        
        elif manoUsuario == 3:  # Usuario elige lagarto
            if manoCompu == 0:  # Piedra gana a lagarto
                print("Gana compu")
            elif manoCompu == 1:  # Lagarto gana a papel
                print("Gana usuario")
            elif manoCompu == 2:  # Tijeras gana a lagarto
                print("Gana compu")
            else:  # Lagarto gana a Spock
                print("Gana usuario")
        
        elif manoUsuario == 4:  # Usuario elige Spock
            if manoCompu == 0:  # Spock gana a piedra
                print("Gana usuario")
            elif manoCompu == 1:  # Papel gana a Spock
                print("Gana compu")
            elif manoCompu == 2:  # Spock gana a tijeras
                print("Gana usuario")
            else:  # Lagarto gana a Spock
                print("Gana compu")

    print("Mano compu " + str(manoCompu))

else:
    print("Mano inválida, manazo!")
```
