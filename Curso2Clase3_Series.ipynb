# nlargest(N) y nsmallest(N)

Los mas grandes y los mas chicos

```python
houses = pd.read_csv("/home/tiamat/Desktop/Python_Dojo/data/kc_house_data.csv")
houses
```

```python
houses['price'].nlargest(10)
#Las primeras 10 casas mas caras :0
```

```python
houses['price'].nsmallest(10)
#Las primeras 10 casas mas baratas :0
```

```python
houses['price'].nlargest(10).sum() #oneliners, una linea que hace muchas cosas al mismo tiempo
#Podemos ademas sumar los precios de las casas mas caras
```

```python
#Que pasa que tenemos datos que se repiten?
#Por ejemplo, tenemos casas con el mismo precio
#Fijate aqui que en vez de houses['price'] usamos house.price

houses.price.nsmallest(20, keep = "all") #Aqui quiero ver todas las que tienen el mismo precio?
```

```python
houses.price.nsmallest(20, keep = "first") #Aqui quiero ver solamente las primeras que tienen ese precio
```

```python
houses.price.nsmallest(20, keep = "last") #Aqui quiero ver solamente las primeras que tienen ese precio
#last in occurrence
```
