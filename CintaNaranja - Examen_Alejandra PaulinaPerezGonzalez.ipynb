Para convertir grados Celcius a Farenheit la frmula es:

Fahrenheit = Celsius * 9/5 + 32

```python
print('Hola, usuario. Inserte una temperatura en grados celsius y te la devolvere en Farenheit. No pongas otra cosa que no son numeros')
celsius =  input('Temperatura en *C')
farenheit = (float(celsius) * 9/5) + 32
print('La temperatura ' + str(celsius) + ' *C es igual a ' + str(farenheit) + ' *F' )
```
