```python
#Ciclo for que me de los numeros pares

x = range(11)
# Como sabemos si un numero es par o non?
for a in x:
    if(a%2 ==0):
        print(a)
```

```python
def numeros_pares(N):
    for i in range(N):
        if(i%2 == 0):
            print(i)
```

numeros_pares(10)

```python
def numeros_pares_2(N):
    for i in range(0, N, 2):
        print(i)
```

```python
numeros_pares_2(10)
```

Otra cosa que podemos pasarle a read_csv son los nombres de las columnas, Podemos decirle que columnas queremos en nuestro dataset,
Para esto tenemos que especificar header = 0 para que la sustituya

```python
state_pops = pd.read_csv("/home/tiamat/Desktop/Python_Dojo/data/nst-est2020.csv")
state_pops.head
```

```python
#Crear una lista de nombres en minuscula

column_names = state_pops.columns
column_names
```

```python
#Las puedo converttir en minusculas
lista = []
for cn in column_names:
    lista.append(cn.lower())
lista
```

```python
#Otra manera de hacerlo

lista = [cn.lower() for cn in column_names]
lista
```

```python
#con esto ya podemos asignarlo a nusetro dataframe
```

```python
state_pops.columns = lista
state_pops
```
